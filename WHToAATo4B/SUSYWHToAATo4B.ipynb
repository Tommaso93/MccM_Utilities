```python
gridpacks_map = {'2016': gridpacks_path_pre2017,
                 '2017': gridpacks_path_2017,
                 '2018': gridpacks_path_2017}

fragment_map = { '2016': fragment_pre2017,
                 '2017': fragment_2017,
                 '2018': fragment_2018}

datasetName_map = { '2016': datasetName_tmp_pre2017,
                    '2017': datasetName_tmp_2017,
                    '2018': datasetName_tmp_2017}

MadgraphVersion = re.compile("V5_2\.[0-9]\.[0-9]")


conditions = '2018'
with open('WH_AA_4B_'+conditions+'.csv', 'w') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    csvwriter.writerow(['Dataset name','Events', 'fragment','notes','Generator'])
    
    
    for mass, path in gridpacks_map[conditions].items():
            #print('\t %s - %s' %(key.replace('VALUE',value),path))
            version = MadgraphVersion.search(path).group(0)
            dataset_name = datasetName_map[conditions].replace('MASSVALUE',mass)
            events=1000000
            final_fragment = fragment_map[conditions].replace('__GRIDPACK__',path)
            note = dataset_name.replace('_',' ')
            generators="Madgraph_"+str(version)+" Pythia8"
            csvwriter.writerow([dataset_name, events, final_fragment, note, generators])
```
